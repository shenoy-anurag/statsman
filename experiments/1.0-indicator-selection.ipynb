```python
records = subset.to_records()
```

```python
records
```

```python
import json

grouped_data = {}

# We extract the year strings dynamically from your columns list (skipping the first 4 metadata columns)
years = columns[4:]

for record in records:
    series_name = record[1]
    series_code = record[2]
    country_name = record[3]  # You can inject this below if you want name metadata!
    country_code = record[4]

    if country_code in ['MAC', 'HKG']:
        continue
    
    # 1. Initialize the Series block if it's the first time we've encountered it
    if series_code not in grouped_data:
        grouped_data[series_code] = {
            "indicatorName": series_name,
            "data": {}
        }
        
    # 2. Iterate through all 50+ trailing years for this specific country
    year_timeline = {}
    for i, year in enumerate(years):
        val = record[5 + i] # Offsetting by 5 to jump past the ID & metadata strings
        
        # Clean '..' blank strings to actual Python None (JSON null)
        if val == '..' or not val:
            parsed_val = None
        # Convert numeric strings into floats/ints so they graph properly
        elif str(val).replace('.', '', 1).isdigit() or str(val).replace('-', '', 1).replace('.', '', 1).isdigit():
            parsed_val = float(val) if '.' in str(val) else int(val)
        else:
            parsed_val = val
            
        year_timeline[year] = parsed_val
        
    # 3. Attach this country's timeline inside the parent Series block
    grouped_data[series_code]["data"][country_code] = year_timeline


# Dump to cleanly indented JSON 
with open('../data/world-development-indicators-grouped.json', 'w') as f:
    json.dump(grouped_data, f, indent=2)
    
print(f"Successfully grouped {len(records)} records into {len(grouped_data.keys())} indicator categories!")
```
