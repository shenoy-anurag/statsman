```python
import json
import pandas as pd
```

```python
cap_df = pd.read_csv("../data/data-gov-india/electricity-installed-capacity.csv")
```

```python
columns = cap_df.columns.tolist()[1:]
```

```python
# cap_df.pivot(columns=['Parameter'])
cap_df = cap_df.T
cap_df
```

```python
cap_df.replace('-', 0, inplace=True)
```

```python
cap_df.columns = cap_df.iloc[0]
cap_df = cap_df[1:]
```

```python
# cap_df.set_index(keys='Parameter')
print(cap_df.columns.tolist())
# print(cap_df.index)
# cap_df['Year'] = cap_df.index
# cap_df
```

```python
solar_columns = [
    "Solar - Ground Mounted (in MW)",
    "Solar - Roof-top (in MW)",
    "Solar - Hybrid-solar (in MW)",
    "Solar - Off-grid (in MW)",
]
biomass_columns = [
    "Bio-power - Bagasse (in MW)",
    "Bio-power - Non-bagasse (in MW)",
    "Bio-power - Waste-to-energy (in MW)",
    "Bio-power - Waste-to-energy-off-grid (in MW)",
]
```

```python
cap_df.fillna(0, inplace=True)

for col in cap_df.columns:
    cap_df[col] = cap_df[col].astype(dtype=float)
    
# cap_df['Coal (in MW)'] = cap_df['Coal (in MW)'].astype(dtype=float)
# cap_df['Growth Rate (in %)'] = cap_df['Growth Rate (in %)'].astype(dtype=float)
# cap_df['Hydro (in MW)'] = cap_df['Coal (in MW)'].astype(dtype=float)
```

```python
cap_df['Solar (in MW)'] = cap_df[solar_columns[0]] + cap_df[solar_columns[1]] + cap_df[solar_columns[2]] + cap_df[solar_columns[3]]
cap_df['Bio Power (in MW)'] = cap_df[biomass_columns[0]] + cap_df[biomass_columns[1]] + cap_df[biomass_columns[2]] + cap_df[biomass_columns[3]]
cap_df['RES-Total'] = cap_df['Hydro (in MW)'] + cap_df['Solar (in MW)'] + cap_df['Wind (in MW)'] + cap_df['Bio Power (in MW)']
```

```python
cap_df.rename(columns={
    'Coal (in MW)': 'Coal', 
    'Oil & Gas (in MW)': 'Oil & Gas', 
    'Nuclear (in MW)': 'Nuclear', 
    'Hydro (in MW)': 'Hydro', 
    'Solar (in MW)': 'Solar', 
    'Solar - Ground Mounted (in MW)': 'Solar - Ground Mounted', 
    'Solar - Roof-top (in MW)': 'Solar - Roof-top', 
    'Solar - Hybrid-solar (in MW)': 'Solar - Hybrid-solar', 
    'Solar - Off-grid (in MW)': 'Solar - Off-grid', 
    'Wind (in MW)': 'Wind', 
    'Small-Hydro (in MW)': 'Small-Hydro', 
    'Bio Power (in MW)': 'Bio Power', 
    'Bio-power - Bagasse (in MW)': 'Bio-power - Bagasse', 
    'Bio-power - Non-bagasse (in MW)': 'Bio-power - Non-bagasse', 
    'Bio-power - Waste-to-energy (in MW)': 'Bio-power - Waste-to-energy', 
    'Bio-power - Waste-to-energy-off-grid (in MW)': 'Bio-power - Waste-to-energy-off-grid', 
    }, inplace=True)
```

```python
data = cap_df.to_dict(orient='index')
data
```

```python
cap_df.head()
```

```python
# Dump to cleanly indented JSON
with open("../data/electricity-installed-capacity.json", "w") as f:
    json.dump({'indicator': 'Power Installed Capacity', 'unit': 'MW', 'data': {'IND': data}}, f, indent=2)
```

```python
gen_df = pd.read_csv("../data/data-gov-india/electricity-power-generation.csv")
gen_df
```

```python
gen_df = gen_df.T

gen_df.replace('-', 0, inplace=True)

gen_df.columns = gen_df.iloc[0]
gen_df = gen_df[1:]

gen_df.fillna(0, inplace=True)

for col in gen_df.columns:
    gen_df[col] = gen_df[col].astype(dtype=float)

display(gen_df)
```

```python
gen_df.columns.tolist()
```

```python
hydro_generation_columns = [
    "Hydro - Generation (in MU)",
    "Small-Hydro - Generation (in MU)",
]
# gen_df['Solar (in MU)'] = gen_df[solar_columns[0]] + gen_df[solar_columns[1]] + gen_df[solar_columns[2]] + gen_df[solar_columns[3]]
# gen_df['Bio Power (in MU)'] = gen_df[biomass_columns[0]] + gen_df[biomass_columns[1]] + gen_df[biomass_columns[2]] + gen_df[biomass_columns[3]]
gen_df["Hydro - Generation (in MU)"] = (
    gen_df[hydro_generation_columns[0]] + gen_df[hydro_generation_columns[1]]
)
gen_df["RES-Total - Generation"] = (
    gen_df["Hydro - Generation (in MU)"]
    + gen_df["Solar - Generation (in MU)"]
    + gen_df["Wind - Generation (in MU)"]
    + gen_df["Bio Power - Generation (in MU)"]
)

gen_df.columns = [
    "Coal",
    "Oil & Gas",
    "Nuclear",
    "Hydro",
    "Solar",
    "Wind",
    "Small-Hydro",
    "Bio Power",
    "RES-Total",
    "Total",
    "Growth Rate (in %)",
]
```

```python
gen_df.head(2)
```

```python
gen_df.drop(columns=['Small-Hydro'], inplace=True)

gen_data = gen_df.to_dict(orient='index')
gen_data
```

```python
# Dump to cleanly indented JSON
with open("../data/electricity-power-generation.json", "w") as f:
    json.dump({'indicator': 'Power Generation', 'unit': 'MU', 'data': {'IND': gen_data}}, f, indent=2)
```
