```python
import json
import pandas as pd
```

```python
df = pd.read_csv("../data/terrorism-deaths.csv")
```

```python
df.head()
```

```python
countries_of_interest = [
    "United States",
    "United Kingdom",
    "China",
    "Germany",
    "India",
    "Japan",
    "France",
    "Italy",
    "Russian Federation",
    "Russia",
    "Canada",
    "Australia",
    "Bangladesh",
    "Pakistan",
]
```

```python
subset = df[df['Entity'].isin(countries_of_interest)]
```

```python
records = subset.to_records()

subset['Entity'].unique()
```

```python
records[:5], len(records)
```

```python
# Entity	Code	Year	Fatalities	World region according to OWID
grouped_data = {}

# We extract the year strings dynamically from your columns list (skipping the first 4 metadata columns)
for record in records:
    country_name = record[1]
    country_code = record[2]
    year = record[3]
    fatalities = record[4]
    # print(country_code, year, fatalities)

    if "terrorism_deaths" not in grouped_data:
        grouped_data["terrorism_deaths"] = {"indicatorName": "Terrorism Deaths, 2021", "data": {}}

    if country_code not in grouped_data["terrorism_deaths"]["data"]:
        grouped_data["terrorism_deaths"]["data"][country_code] = {}
    
    if year not in grouped_data["terrorism_deaths"]["data"][country_code]:
        grouped_data["terrorism_deaths"]["data"][country_code][str(year)] = int(fatalities)
    
    grouped_data["terrorism_deaths"]["data"][country_code][str(year)] = int(fatalities)


# Dump to cleanly indented JSON
with open("../data/terrorism-deaths-grouped.json", "w") as f:
    json.dump(grouped_data, f, indent=2)

print(
    f"Successfully grouped {len(records)} records into {len(grouped_data.keys())} indicator categories!"
)
```
